```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
%matplotlib inline

```

```python
df = pd.read_csv("Salary_Data.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.plot(x="YearsExperience", y="Salary", kind="scatter")
```

```python

X_train, X_test, y_train, y_test = train_test_split(df.YearsExperience, df.Salary, test_size=0.2, random_state=10)
```

```python
LR = LinearRegression()
LR.fit(X_train.values.reshape(-1,1), y_train.values.reshape(-1,1))
```

```python
y_pred = LR.predict(X_test.values.reshape(-1,1))
```

```python
from sklearn.metrics import r2_score , mean_squared_error, mean_absolute_error
```

```python
mean_squared_error(y_test, y_pred)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred
                          ))
```

```python
mean_absolute_error(y_test, y_pred
                   )
```

```python
r2_s = r2_score(y_test, y_pred)

```

```python
y_test.shape

```

```python
1 - ((1-r2_s)*(y_test.shape)[0]/ (
    y_test.shape[0]-1-1))
```

```python
LR.coef_
```

```python
y_pred = LR.predict(X_train.values.reshape(-1, 1))

plt.scatter(X_train, y_train, color='red', label='Actual Data')
plt.plot(X_train, y_pred, color='blue', label='Best Fit Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Years of Experience vs. Salary with Best Fit Line')
plt.legend()
plt.show()
```
